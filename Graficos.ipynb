```python
df_train = pd.read_csv('housing_train.csv')
```

```python
plt.scatter(df_train['tasa_crimen'],df_train['valor'])    ###Grafico de correlacion
```

```python
sns.pairplot(df_train, x_vars= ['tasa_crimen','lz','cuartos','rio'], y_vars= ['valor'])   ##Creacion de multiples graficos de dispersion
```

```python
sns.pairplot(df_train, x_vars= ['tasa_crimen','lz','cuartos','rio'], y_vars= ['valor'], kind="reg") ##Multiples graficos con la linea de regresion trazada
```

```python
sns.heatmap(df_train.corr())  ##Mapa de calor para saber q variables tienen la mayor correlacion entre si
```

```python
sns.boxplot(x= df_train['autopistas'], y= df_train['valor'])   ##Graficos de cajas
```

```python
sns.histplot(df_train["cuartos"])   ##Histograma
```

```python
sns.histplot(df_train["lz"])
```
